# Stochastic Simulation of a Limit Order Book

This stochastic model was inspired by [Cont, Stoikov, and Talreja (2010)](http://rama.cont.perso.math.cnrs.fr/pdf/CST2010.pdf).

```python
from src.simulation import Simulation
from src.exchange import Exchange
import matplotlib.pyplot as plt
import numpy as np
```

Parameters for the stochastic order book simulation. The limit order arrival rates are listed in order of distance to the opposite best quote.


```python
initial_bid_order_depths = {5000: 15, 4999: 10, 4998: 5}
initial_ask_order_depths = {5001: 15, 5002: 10, 5003: 5}
market_order_rate = 1.2
limit_order_rates = [0.54, 0.29, 0.25, 0.2, 0.1]
length_of_simulation = 1000
price_levels = 10000
```

Run the simulation.

```python
ex = Exchange(price_levels)
sim = Simulation(ex, limit_order_rates, market_order_rate, initial_bid_order_depths, initial_ask_order_depths)

for i in range(length_of_simulation):
    sim.step()
    ex.step(i)
```

Plot the mid price over time.

```python
plt.plot(np.arange(length_of_simulation), ex.get_mid_price_history())
plt.show()
```
